```python
# ## 3. Formula API - Time-Series Endpoint 
# In this example, we will retrieve the last 6 quarters of sales and price and the long security name for Apple, IBM, and Microsoft. We will also set the calendar to return data on the dates that correspond to the United States trading days.

# Get Data
with fds.sdk.Formula.ApiClient(configuration) as api_client:
    # Create Instance
    api_instance = TimeSeriesApi(api_client)

    # Request Object to Define Parameters
    time_series_request = TimeSeriesRequest(
        data=TimeSeriesRequestData(
            ids = [
                "IBM-US",
                "MSTR-US",
                "NVDA-US",
            ],
            formulas = [
                "PROPER_NAME",
                "P_PRICE(0,-5,Q)",
                "FF_SALES(QTR,0,-5,Q,,USD)"
            ],
            calendar = "NAY",
            flatten = "Y"
        ),
    )

    # Send Request
    time_series_response_wrapper = api_instance.get_time_series_data_for_list(time_series_request)
    time_series_response = time_series_response_wrapper.get_response_200()

    # Convert to Pandas Dataframe
    time_series_results = pd.DataFrame(time_series_response.to_dict()['data'])
    display(time_series_results)
    
```
